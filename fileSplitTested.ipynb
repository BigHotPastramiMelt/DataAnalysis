```python
#explore store_restaurant, use pandas since file is not large
store_restaurant = pd.read_csv("C:\\Users\\patrick.carey\\Desktop\\wcai\\storeRestaurant.csv")
store_restaurant
```

```python
#retrieve chicago FAF codes and turn into list
Chicago_FAF = store_restaurant.loc[store_restaurant['FAF_MKT_NM'] == 'CHICAGO']
Chicago_FAF_codes = Chicago_FAF["STORE_NUMBER"]
Chicago_FAF_codes = list(Chicago_FAF_codes)
NY_FAF = store_restaurant.loc[store_restaurant['FAF_MKT_NM'] == 'NEW YORK-NEW JERSEY']
NY_FAF_codes = NY_FAF["STORE_NUMBER"]
NY_FAF_codes = list(NY_FAF_codes)
LA_FAF = store_restaurant.loc[store_restaurant['FAF_MKT_NM'] == 'LOS ANGELES']
LA_FAF_codes = LA_FAF["STORE_NUMBER"]
LA_FAF_codes = list(LA_FAF_codes)
Houston_FAF = store_restaurant.loc[store_restaurant['FAF_MKT_NM'] == 'HOUSTON']
Houston_FAF_codes = Houston_FAF["STORE_NUMBER"]
Houston_FAF_codes = list(Houston_FAF_codes)
```

```python
#split up pos_ordersale
with open("C:\\Users\\patrick.carey\\Downloads\\pos_ordersale_full.csv", 'r') as f:
    with open("C:\\Users\\patrick.carey\\Desktop\\wcai\\pos_ordersale_FAF_NY.csv", "a") as NY: #creates file to write to
        with open("C:\\Users\\patrick.carey\\Desktop\\wcai\\pos_ordersale_FAF_LA.csv", "a") as LA:
            with open("C:\\Users\\patrick.carey\\Desktop\\wcai\\pos_ordersale_FAF_CHICAGO.csv", "a") as CHI:
                with open("C:\\Users\\patrick.carey\\Desktop\\wcai\\pos_ordersale_FAF_HOUSTON.csv", "a") as HOUS:
                    with open("C:\\Users\\patrick.carey\\Desktop\\wcai\\pos_ordersale_FAF_SF.csv", "a") as SF:
                        line = f.readline() #read variable headings
                        lineList = line.split(",")
                        print(line)
                        keyCol = 'STORE_SAT_NUMBER'
                        keyCol = lineList.index(keyCol)
                        NY.write(line) #write variable headings to new file
                        LA.write(line)
                        CHI.write(line)
                        HOUS.write(line)
                        SF.write(line)
                        while line:
                            if lineList[keyCol] in NY_FAF_codes:
                                NY.write(line)
                            elif lineList[keyCol] in LA_FAF_codes:
                                LA.write(line)
                            elif lineList[keyCol] in Chicago_FAF_codes:
                                CHI.write(line)
                            elif lineList[keyCol] in Houston_FAF_codes:
                                HOUS.write(line)
                            else:
                                SF.write(line)
                            line=f.readline()
                            lineList=line.split(",")
                            
```
