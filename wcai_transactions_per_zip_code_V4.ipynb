```python
# combine cash and vps samples 
pos_ordersale_smp = pd.concat([pos_ordersale_vps_customer_smp, pos_ordersale_cash_customer_smp], axis=0)
pos_ordersale_smp.shape
```

```python
# filter out non-Chicago data 
pos_ordersale_smp_chicago= pos_ordersale_smp[pos_ordersale_smp['STORE_SAT_NUMBER'].isin(chicago_store_numbers_list)]
pos_ordersale_smp_chicago.shape  
```

```python
pos_ordersale_smp_chicago.dtypes
```

```python
#pos_ordersale_smp_chicago['date']
```

```python
#pos_ordersale_smp['date'] = pd.to_datetime(pos_ordersale_smp['date'])
pos_ordersale_smp_chicago['date'] = pd.to_datetime(pos_ordersale_smp_chicago['date'].astype(str), format='%Y%m%d')
```

```python
pos_ordersale_smp_chicago.dtypes
```

```python
# sort the date column
pos_ordersale_smp_chicago = pos_ordersale_smp_chicago.sort('date')
pos_ordersale_smp_chicago.head()
```

```python
# Number of transactions by store
num_trans_bystore = pos_ordersale_smp_chicago['STORE_SAT_NUMBER'].value_counts(sort=False)
num_trans_bystore
```

```python
#change into dataframe and sort
num_trans_bystore2=pd.DataFrame(num_trans_bystore)
num_trans_bystore2.reset_index(level=0, inplace=True)
num_trans_bystore2.columns = ['STORE_NUMBER', '#TRANSACTIONS']
num_trans_bystore2 = num_trans_bystore2.sort('STORE_NUMBER')
num_trans_bystore2
```

```python
#create new df containing only store number, city, and zip
chicago_stores2 = chicago_stores[['STORE_NUMBER','STORE_CITY','STORE_ZIP']]
#df2 = df1[['a','b']]
chicago_stores2 = chicago_stores2.sort('STORE_NUMBER')
chicago_stores2.head()
```

```python
#combine the last two dataframes together on store_number
result = pd.merge(chicago_stores2, num_trans_bystore2, on='STORE_NUMBER', )
result = result.sort('STORE_NUMBER')
result
```

```python
#result.to_csv("chicago_num_transactions.csv")
```

```python
#group by store zip, adding up the number of transactions in each zip code
result_grouped = result.groupby('STORE_ZIP')['#TRANSACTIONS'].sum()
#g1 = df1.groupby( [ "Name", "City"] ).count()
result_grouped
```

```python
#convert to df and sort
result_grouped2=pd.DataFrame(result_grouped)
result_grouped2.reset_index(level=0, inplace=True)
result_grouped2.columns = ['ZIP_CODE', 'TOTAL_TRANSACTIONS']
result_grouped2 = result_grouped2.sort('ZIP_CODE')
result_grouped2
```

```python
#write to csv
result_grouped2.to_csv("chicago_num_transactions2.csv")
#this file was then manually combined to another excel file i have with the counts of stores in each zip code
#to get the number of transactions per store, for each zip code. the file was then plotted in tableau.
```

```python
# total tax amouont per store
total_tax_by_store = pos_ordersale_smp_chicago.groupby('STORE_SAT_NUMBER')['TaxAmount'].sum()
total_tax_by_store
```

```python
#convert to df and sort
total_tax_by_store2=pd.DataFrame(total_tax_by_store)
total_tax_by_store2.reset_index(level=0, inplace=True)
total_tax_by_store2.columns = ['STORE_NUMBER', 'TOTAL_TAX_AMOUNT']
#result_grouped2 = result_grouped2.sort('ZIP_CODE')
total_tax_by_store2
```

```python
#group by store zip, adding up the total taxes of all the stores in each zip code
total_tax_by_zip = total_tax_by_store3.groupby('STORE_ZIP')['TOTAL_TAX_AMOUNT'].sum()
#total_tax_by_zip
#convert to df and sort
total_tax_by_zip2=pd.DataFrame(total_tax_by_zip)
total_tax_by_zip2.reset_index(level=0, inplace=True)
total_tax_by_zip2.columns = ['ZIP_CODE', 'TOTAL_TAX_AMOUNT']
total_tax_by_zip2 = total_tax_by_zip2.sort('ZIP_CODE')
total_tax_by_zip2.head()
```

```python
#add on the number of stores per zip
total_tax_by_zip3 = pd.merge(total_tax_by_zip2, chicago_store_count2, on='ZIP_CODE', )
total_tax_by_zip3 = total_tax_by_zip3.sort('ZIP_CODE')
total_tax_by_zip3.head()
```

```python
#add on the number of transactions per zip
total_tax_by_zip4 = pd.merge(total_tax_by_zip3, result_grouped2, on='ZIP_CODE', )
total_tax_by_zip4 = total_tax_by_zip4.sort('ZIP_CODE')
total_tax_by_zip4.head()
```

**Number of Complaints by Zip Code**

```python
complaints = pd.read_csv('complaint_data.csv')
complaints.head(3)
# has not yet filtered out non-Chicago data 
```

```python
complaints.shape
#has not yet filtered out non-Chicago data
```

```python
# filter out non-Chicago data 
complaints_chicago= complaints[complaints['STORE'].isin(chicago_store_numbers_list)]
complaints_chicago.shape 
```

```python
# Number of complaints by store
num_complaints_bystore = complaints_chicago['STORE'].value_counts(sort=True)
num_complaints_bystore
```

```python
#change into dataframe and sort
num_complaints_bystore2=pd.DataFrame(num_complaints_bystore)
num_complaints_bystore2.reset_index(level=0, inplace=True)
num_complaints_bystore2.columns = ['STORE_NUMBER', '#COMPLAINTS']
num_complaints_bystore2 = num_complaints_bystore2.sort('STORE_NUMBER')
num_complaints_bystore2.head()
```

```python
#combine the num_complaints_bystore2 with chicago_stores2 to be able to get number of complaints per zip
num_complaints_bystore3 = pd.merge(chicago_stores2, num_complaints_bystore2, on='STORE_NUMBER', )
num_complaints_bystore3 = num_complaints_bystore3.sort('STORE_NUMBER')
num_complaints_bystore3.head()
```

```python
#group by store zip, adding up the total complaints of all the stores in each zip code
total_complaints_by_zip = num_complaints_bystore3.groupby('STORE_ZIP')['#COMPLAINTS'].sum()
#total_complaints_by_zip
#convert to df and sort
total_complaints_by_zip2=pd.DataFrame(total_complaints_by_zip)
total_complaints_by_zip2.reset_index(level=0, inplace=True)
total_complaints_by_zip2.columns = ['ZIP_CODE', 'TOTAL_COMPLAINTS']
total_complaints_by_zip2 = total_complaints_by_zip2.sort('ZIP_CODE')
total_complaints_by_zip2.head()
```

```python
#add on the number of stores per zip
total_complaints_by_zip3 = pd.merge(total_complaints_by_zip2, chicago_store_count2, on='ZIP_CODE', )
total_complaints_by_zip3 = total_complaints_by_zip3.sort('ZIP_CODE')
total_complaints_by_zip3.head()
```

```python
total_complaints_by_zip3['Comps_perStore']=(total_complaints_by_zip3.TOTAL_COMPLAINTS)/(total_complaints_by_zip3.COUNT)
```
