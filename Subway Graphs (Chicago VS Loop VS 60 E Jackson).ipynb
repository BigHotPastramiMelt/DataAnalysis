# Chicago 2014 Transactions

```python
pos_ordersale_2014Chicago = pd.read_csv('pos_ordersale_2014Chicago.csv')
pos_ordersale_2014Chicago.head(4)
```

```python
pos_ordersale_2014Chicago.shape
```

```python
pos_ordersale_2014Chicago.dtypes
```

```python
pos_ordersale_2014Chicago['date'] = pd.to_datetime(pos_ordersale_2014Chicago['date'])
```

```python
pos_ordersale_2014Chicago.dtypes
```

```python
pos_ordersale_2014Chicago.head(4)
```

```python
# Number of transactions by store
num_trans_bystore = pos_ordersale_2014Chicago['STORE_SAT_NUMBER'].value_counts(sort=False)
num_trans_bystore
```

```python
# number of stores by date 
pos_ordersale_2014Chicago_groupbydate=pos_ordersale_2014Chicago.groupby('date')
pos_ordersale_2014Chicago_num_stores_bydate=pos_ordersale_2014Chicago_groupbydate['STORE_SAT_NUMBER'].nunique()
pos_ordersale_2014Chicago_num_stores_bydate
```

```python
pos_ordersale_2014Chicago_num_stores_bydate.plot()
plt.xlabel('Date')
plt.ylabel('# of Stores')
plt.title('# of Stores in Chicago (with >= 1 transaction) by date')
```

```python
pos_ordersale_2014Chicago['date'].value_counts(sort=False).plot(grid=True)
plt.xlabel('Date')
plt.ylabel('# of Transactions')
plt.title('# of Subway Transactions in Chicago in 2014 by date')
```

The graph above shows that there's an increase in the number of total Chicago transactions from January to December 2014. However, this graph doesn't take into account the number of stores in the POS system.

By dividing the # of transactions by the # of stores in the POS system on a given day, the graph below shows the average # of transactions in a Chicago store over the 2014 period. 

```python
(pos_ordersale_2014Chicago['date'].value_counts(sort=False)/pos_ordersale_2014Chicago_num_stores_bydate).plot(grid=True)
plt.xlabel('Date')
plt.ylabel('# of Transactions')
plt.title('Average # of Subway Transactions Per Store in Chicago in 2014 by date')
```

The graph shows that in 2014 in Chicago, stores averaged 100-250 transactions daily. The trend does show some improvement when comparing the first half of the year to the second half of the year.
