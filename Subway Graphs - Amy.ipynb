# Chicago 2014 Transactions (sample)

```python
pos_ordersale_cash_customer_smp2014 = pd.read_csv('pos_ordersale_cash_customer_smp2014.csv')
pos_ordersale_cash_customer_smp2014.head(4)
# has not yet filtered out non-Chicago data 
```

```python
pos_ordersale_cash_customer_smp2014.shape
# has not yet filtered out non-Chicago data 
```

```python
pos_ordersale_vps_customer_smp2014 = pd.read_csv('pos_ordersale_vps_customer_smp2014.csv')
pos_ordersale_vps_customer_smp2014.head(4)
# has not yet filtered out non-Chicago data 
```

```python
pos_ordersale_vps_customer_smp2014.shape
# has not yet filtered out non-Chicago data 
```

```python
# combine cash and vps 2014 samples 
pos_ordersale_2014smp = pd.concat([pos_ordersale_vps_customer_smp2014, pos_ordersale_cash_customer_smp2014], axis=0)
pos_ordersale_2014smp.shape
```

```python
pos_ordersale_2014smp.dtypes
```

```python
pos_ordersale_2014smp['date']= pd.to_datetime(pos_ordersale_2014smp['date'])
```

```python
pos_ordersale_2014smp.dtypes
```

```python
# filter out non-Chicago data 
pos_ordersale_2014smp_chicago= pos_ordersale_2014smp[pos_ordersale_2014smp['STORE_SAT_NUMBER'].isin(chicago_store_numbers_list)]
pos_ordersale_2014smp_chicago.shape                                                                      
```

```python
pos_ordersale_2014smp_chicago.head(4)
```

```python
# sort the date column
pos_ordersale_2014smp_chicago = pos_ordersale_2014smp_chicago.sort('date')
pos_ordersale_2014smp_chicago.head()
```

```python
# Number of transactions by store
num_trans_bystore = pos_ordersale_2014smp_chicago['STORE_SAT_NUMBER'].value_counts(sort=False)
num_trans_bystore
```

```python
# number of stores by date 
pos_ordersale_2014smp_chicago_groupbydate=pos_ordersale_2014smp_chicago.groupby('date')
pos_ordersale_2014smp_chicago_num_stores_bydate=pos_ordersale_2014smp_chicago_groupbydate['STORE_SAT_NUMBER'].nunique()
pos_ordersale_2014smp_chicago_num_stores_bydate
```

```python
pos_ordersale_2014smp_chicago['date'].value_counts(sort=False).plot(grid=True)
plt.xlabel('Date')
plt.ylabel('# of Transactions')
plt.title('# of Subway Transaction in Chicago in 2014 by date(sample)')
```

The graph above shows that there's an increase in the number of total Chicago transactions from January to December 2014. However, this graph doesn't take into account the number of stores in the POS system.

By dividing the # of transactions by the # of stores in the POS system on a given day, the graph below shows the average # of transactions in a Chicago store over the 2014 period. 

```python
(pos_ordersale_2014smp_chicago['date'].value_counts(sort=False)/pos_ordersale_2014smp_chicago_num_stores_bydate).plot(grid=True)
plt.xlabel('Date')
plt.ylabel('# of Transactions')
plt.title('Average # of Subway Transaction Per Store in Chicago in 2014 by date(sample)')
```

The graph shows that in 2014 in Chicago, stores averaged 1-2 transactions daily. The trend does show some improve when comparing the first half of the year (stores averaged 1-2 transactions each day) to the second half of the year (stores averaged 1.5-2 transactions each day).
